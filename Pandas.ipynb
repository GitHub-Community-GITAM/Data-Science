## Creating data using pandas

There are two core objects in pandas: the <b>DataFrame</b> and the <b>Series</b>.

### Creating a DataFrame

A DataFrame is a table. It contains an array of individual entries, each of which has a certain value. Each entry corresponds to a row (or record) and a column.

Let's take an example. Jack and Jill are discussing about sweet and savoury food. Jack prefers sweet more than savoury and Jill prefers the opposite. Lets create a DataFrame to depict their likes and dislikes.

```python
pd.DataFrame({'Jack':["Very yum","Awful"],'Jill':["Not my type",'I loved it']})
```

Here, we cannot really know what Jack and Jill is talking about. Let's change the indexes from 0,1 to something more informative

```python
pd.DataFrame({'Jack':["Very yum","Awful"],'Jill':["Not my type",'I loved it']},index=["Chocolates","Chips"])
```

### Creating Series

A Series, by contrast, is a sequence of data values. If a DataFrame is a table, a Series is a list. And in fact you can create one with nothing more than a list:

```python
pd.Series([1,2,3,4])
```

Let's make it more specific. Lets print a list of items and its available number sold by Grocery A.

```python
pd.Series([30,35,29],index=["Chocolates","Chips","Biscuits"],name='Grocery A')
```

## Indexing in Pandas

### 1)iloc - index-based selection

iloc involves selecting data based on its numerical position in data

To select the first row of data in a DataFrame, we use:

```python
net_titles.iloc[0]
```

```python
net_titles.iloc[:,1]
```

```python
net_titles.iloc[:3,[1,2,3]]
```

### 2) loc - label-based selection

In this paradigm, it's the data index value, not its position, which matters.

For example, to get the first entry in reviews, we would now do the following:

```python
net_titles.loc[0,'title']
```

```python
net_titles.loc[:,['title','director']]
```

## Conditional Selection

To check which of the shows were created in the United States, we can do the following:

```python
net_titles.country == 'United States'
```

Now to get the name and details of the shows created in the Unites States, we can do the following

```python
net_titles.loc[net_titles.country=="United States"]
```

Inbuilt  conditional selectors in pandas
- isin
- isnull or notnull

isin is lets you select data whose value "is in" a list of values. 

```python
net_titles.loc[net_titles.country.isin(['United States', 'India'])]
```

```python
net_titles.loc[net_titles.cast.notnull()]
```

```python
net_titles.loc[net_titles.cast.isnull()]
```

## Renaming and Combining of data

### Renaming
The first function we'll introduce here is rename(), which lets you change index names and/or column names. For example, to change the 'duration' column in our dataset to 'no. of seasons', we would do:

```python
net_titles.rename(columns={'duration' : 'no. of seasons'})
```

<b>rename()</b> lets you rename index or column values by specifying a index or column keyword parameter, respectively. It supports a variety of input formats, but usually a Python dictionary is the most convenient. Here is an example using it to rename some elements of the index.

```python
net_titles.rename(index={0 : 'First',1:'Second'})
```

```python
net_titles.rename_axis('s.no',axis='rows').rename_axis('fields',axis='columns')
```
